Create a program that generates random coordinates for an imaginary rectangle and the player has to guess coordinates and see if it is inside the rectangle or not

<u>Description of the program</u>

- The program displays a message to the user showing the coordinates of a rectangle.
- The user should enter the coordinates of a point

<u>**How many objects do we need for our program to run?**</u>

1. meesage ---> string ---> class that produces string instances (already designed)
2. coordinates ---> float (rectangle and points)
3. rectangle ---> new class ---> create object instance of that class
4. point ---> new class ---> create object instance of that class

<u>**Development Steps (OOP design flow)**</u>

1. writing down the objects
2. write a class for each of that object
    - what are the minimum requirements for that specific class to exist
3. writing methods/functions for each class
4. calling the classes (instantiate) and their methods



````python
# the class

class Point:

    def __init__(self, x : int, y : int):
        self.x = x
        self.y = y

# Creating an object instance of type Point

point1 = Point(2,3)
type(point1)
````

<u>**Point Class**</u>

- `def __init__` ---> holds the parameters of the object I am creating
- the `__init__` function is executed at object creation  ---> constructor
- when we instantiate an object ---> it is of the type Point
- we can create as many objects as we want with the type Point from the class Point
- `__main__.Point`
    - `__main__` is the name of the current script
    - `Point` is the type of the object ---> the class
    - `__main__` is file where the class is located
- `self`
    - variable that holds the object instance that is created by that class
    - `x` will hold `2` and `self.x` will hold `x`
    - `self.x` is called an instance variable
    - `self.x` is the equivalent of `point1.x`




````python
# Creating a method

class Point:

    def __init__(self, x : int, y : int):
        self.x = x
        self.y = y

    def falls_in_rectangle(self, lowleft : tuple, upright : tuple) -> bool:
        if lowleft[0] < self.x < upright[0] \
        and lowleft[1] < self.y < upright[1]:
            return True
        else:
            return False
        
    def distance_from_point(self, point : Point) -> float:
        return ((self.x - point.x)**2 + (self.y - point.y)**2) ** 0.5         

# intantiate the Point object        
point2 = Point(3,4)

# call method 1
print(point2.falls_in_rectangle((1,1), (7,9)))

# call method 2
point3 = Point(3,1)
point2.distance_from_point(point3)
````

<u>**Methods**</u>

- using verbs to name the methods `falls_in_rectagle`
- I need to pass the `self` in the method of that class
